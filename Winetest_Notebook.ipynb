```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("/content/WineQT.csv")
```

```
df
```

```
df.head()
```

```
value_sample = df['quality']
print(value_sample)
```

```
quality_counts = value_sample.value_counts()
print(quality_counts)
```

```
df.describe()
```

```
df.info()
```

```
df.dtypes
```

```
plt.figure(figsize=(8,6))
plt.bar(quality_counts.index, quality_counts,color='LIGHTGREEN')
plt.title("Count Plot Of Quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

```
sns.set_style("darkgrid")
```

```
numerical_columns = df.select_dtypes(include=["int64","float64"]).columns
print(numerical_columns)
```

```
plt.figure(figsize=(12, 12))  # Increase figure size for better visibility
num_cols = len(numerical_columns)
num_rows = (num_cols + 2) // 3  # Calculate the number of rows needed

for i, column in enumerate(numerical_columns, 1):
  plt.subplot(num_rows, 3, i)  # Adjust subplot grid dynamically
  sns.histplot(df[column], kde=True)

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```
sns.swarmplot(x="quality", y="alcohol",data = df, palette= 'viridis')
plt.title("Alcohol Content by Wine Quality")
plt.xlabel("Wine Quality")
plt.ylabel("Alcohol Content")
plt.show()
```

```
sns.set_palette('Pastel1')
plt.figure(figsize=(10, 6))
sns.pairplot(df)
plt.suptitle('Pair Plot of Wine Quality Dataset')
plt.show()
```

```
df['quality'] = df['quality'].astype(str)
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Wine Quality Dataset')
plt.show()
```
